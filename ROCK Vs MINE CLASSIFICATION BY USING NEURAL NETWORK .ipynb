```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import tensorflow as tf 
from tensorflow import keras 
from sklearn.model_selection import train_test_split
```

```python
for i in data:
    print(data[i].isnull().sum())
```

```python
#one hot coding 
data.replace("R", 0, inplace=True)
data.replace("M", 1, inplace=True)
```

```python
data['R'].dtypes
```

```python
data.columns
```

```python
for i,j in enumerate(data):
    #print(i, j)
    data.rename(columns = {j:i}, inplace = True)
```

```python
data.columns
```

```python
data.shape
```

```python
data[60].value_counts()
```

```python
x = data.drop(60, axis="columns")
y = data[60]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60, ), activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model.fit(x_train, y_train, epochs=200)
```

```python
model.evaluate(x_test, y_test)
```

```python
y_predict_test = model.predict(x_test).reshape(-1)
a = np.round(y_predict_test)
a
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(a, y_test))
```

#### dropout layer

```python
model_1 = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60, ), activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])
model_1.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model_1.fit(x_train, y_train, epochs=200)
```

```python
model_1.evaluate(x_test, y_test)
```
